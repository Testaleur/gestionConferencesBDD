##### Imports

```python
import sqlite3
DB_PATH = "../gestionConferences.db"
```

##### Vues à créer

```python
views = [
  """DROP VIEW IF EXISTS v_conference_utilisateur;""",
  """DROP VIEW IF EXISTS v_conference_responsable;""",

  """
  CREATE VIEW IF NOT EXISTS v_conference_utilisateur AS
  SELECT
    id_conference,
    title,
    starting_date,
    ending_date,
    introduction,
    city,
    country,
    series,
    type,
    key_words,
    id_universite,
    associated_conf
  FROM conference;
  """,
  """
  CREATE VIEW IF NOT EXISTS v_conference_responsable AS
  SELECT
    *
  FROM conference;
  """,
  """
  CREATE VIEW IF NOT EXISTS v_conference_admin AS
  SELECT
    *
  FROM conference;
  """
]
```

##### Création des vues

```python
conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

for view_sql in views:
  cursor.execute(view_sql)

conn.commit()
conn.close()
print("✅ Vues conférence créées")
```
